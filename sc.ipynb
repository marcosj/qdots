# Noise in quantum dots
Loss–DiVincenzo (LD) qubit

Singlet–triplet (ST) qubit

```python
ld = { 't2': 21e-6, 't2q': 40e-9, 'tr': 24e-6,
       'p1q': 4e-4, 'p2q': 2e-3, 'pr': 2.4e-3, 'pp': 6.5e-3 } # dq
st = { 't2': 14.8e-6, 't2q': 40e-9, 'tr': 2.4e-6,
       'p1q': 4e-3, 'p2q': 4e-3, 'pr': 4e-4, 'pp': 4e-3 } # aq
p = lambda t, t2: .5 * ( 1 - exp(-t*t/t2/t2) )
```

# Unrotated surface code
c.png

```python
def xxxx(k, D, q):
    # k in S_X
    i, j = k // D, k % D
    # neighbors
    n, w, e, s = k-D, k-1, k+1, k+D
    d = ''
    if i > 0: d = f'rec[-{len(q) - q.index(n)}] ' + d
    if j > 0: d = f'rec[-{len(q) - q.index(w)}] ' + d
    if j < D-1: d = f'rec[-{len(q) - q.index(e)}] ' + d
    if i < D-1: d = f'rec[-{len(q) - q.index(s)}] ' + d
    return d

def noise(S, channel, prob, qubits):
    S.append(f'{channel}({prob}) ' + ' '.join(map(str, qubits)))
    S.append('TICK')

def gen_sc(d, r=1, nm=True):
    D = 2 * d - 1
    # stabs
    x = [D*i+j for i in range(D) for j in range(D) if i % 2 == 0 and j % 2 == 1]
    z = [D*i+j for i in range(D) for j in range(D) if i % 2 == 1 and j % 2 == 0]
    dq = [D*i+j for i in range(D) for j in range(D) if (i + j) % 2 == 0]
    aq = z + x
    # cnots
    xe = [s + o for s in x for o in (0, 1)]
    xw = [s - o for s in x for o in (0, 1)]
    xs = [s + o for s in x for o in (0, D) if s < D*D-D]
    xn = [s - o for s in x for o in (0, D) if s >= D]
    ze = [s + o for s in z for o in (1, 0) if s % D < D-1]
    zw = [s - o for s in z for o in (1, 0) if s % D > 0]
    zs = [s + o for s in z for o in (D, 0)]
    zn = [s - o for s in z for o in (D, 0)]

    # round S
    # applies R_Y(-π/2) to data qubits
    S = ['SQRT_Y_DAG ' + ' '.join(map(str, dq))]
    if nm: noise(S, 'Y_ERROR', ld['p1q'], dq)
    # applies R_Y(+π/2) to ancilla qubits
    S.append('SQRT_Y ' + ' '.join(map(str, aq)))
    if nm: noise(S, 'Y_ERROR', st['p1q'], aq)
    # S_X: applies X stabilizers
    S.append('CZ ' + ' '.join(map(str, xe)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), xe)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in xe])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in xe])
    S.append('CZ ' + ' '.join(map(str, xs)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), xs)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in xs])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in xs])
    S.append('CZ ' + ' '.join(map(str, xn)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), xw)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in xw])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in xw])
    S.append('CZ ' + ' '.join(map(str, xw)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), xw)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in xw])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in xw])
    # applies R_Y(+π/2) to data qubits
    S.append('SQRT_Y ' + ' '.join(map(str, dq)))
    if nm: noise(S, 'Y_ERROR', ld['p1q'], dq)
    # S_Z: applies Z stabilizers
    S.append('CZ ' + ' '.join(map(str, ze)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), ze)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in ze])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in ze])
    S.append('CZ ' + ' '.join(map(str, zs)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), zs)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in zs])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in zs])
    S.append('CZ ' + ' '.join(map(str, zn)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), zn)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in zn])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in zn])
    S.append('CZ ' + ' '.join(map(str, zw)))
    if nm: noise(S, 'PAULI_CHANNEL_2', 14*'0,' + str(st['p2q']), zw)
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], ld['t2']), [q for q in dq if q not in zw])
    if nm: noise(S, 'Z_ERROR', p(st['t2q'], st['t2']), [q for q in aq if q not in zw])
    # applies R_Y(-π/2) to ancilla qubits
    S.append('SQRT_Y_DAG ' + ' '.join(map(str, aq)))
    if nm: noise(S, 'Y_ERROR', st['p1q'], aq)
    # measures ancilla qubits
    if nm: noise(S, 'X_ERROR', st['pr'], aq)
    S.append(f'MR ' + ' '.join(map(str, aq)))
    if nm: noise(S, 'X_ERROR', st['pp'], aq)
    if nm: noise(S, 'Z_ERROR', p(st['tr'], ld['t2']), dq)

    # Preparation
    c = [f'QUBIT_COORDS({j}, {i}) {D*i+j}' for i in range(D) for j in range(D)]
    # prepares data qubits in |+>_L
    c.append('RX ' + ' '.join(map(str, dq)))
    if nm: noise(c, 'Z_ERROR', ld['pp'], dq)
    # prepares ancilla qubits in |0>_L
    c.append('R ' + ' '.join(map(str, aq)))
    if nm: noise(c, 'X_ERROR', st['pp'], aq)

    # Round 0
    c.extend(S)
    c.extend([f'DETECTOR({k%D}, {k//D}, 0) rec[-{len(x)-i}]' for i, k in enumerate(x)])
    c.append('TICK')

    # Rounds 1, ...
    if r > 1:
        c.append(f'REPEAT {r-1} {{')
        c.extend(S)
        c.append('SHIFT_COORDS(0, 0, 1)')
        c.extend([f'DETECTOR({k%D}, {k//D}, 0) rec[-{len(aq)-i}] rec[-{2*len(aq)-i}]'
                 for i, k in enumerate(aq)])
        c.append('TICK')
        c.append('}')

    # Readout
    #if nm: noise(c, 'Z_ERROR', p(ld['tr'], st['t2']), aq)
    if nm: noise(c, 'Z_ERROR', ld['pr'], dq)
    c.append(f'MX ' + ' '.join(map(str, dq)))
    c.extend([f'DETECTOR({k%D}, {k//D}, 1) {xxxx(k, D, dq)}rec[-{len(dq)+len(x)-i}]'
             for i, k in enumerate(x)])

    # X_L
    c.append(f'OBSERVABLE_INCLUDE(0) ' + ' '.join([f'rec[-{(i+1)*d+i*(d-1)}]'
             for i in range(d)]))
    return stim.Circuit('\n'.join(c))
```

```python
circs = {
    (d,r): gen_sc(d, r)
    for d in [3,5,7,9]
    for r in range(1, 19, 1)
}
```

```python
tasks = [
    sinter.Task(
        circuit=circs[(d, r)],
        json_metadata={'d': d, 'r': r},
    )
    for d in [3,5,7,9]
    for r in range(1, 19, 1)
]

stats = sinter.collect(
    num_workers=4,
    tasks=tasks,
    decoders=['pymatching'],
    max_shots=1_000_000,
    max_errors=1_000
)
```

```python
x = [s.json_metadata['r'] for s in stats]
y = [100 - 100 * s.errors / s.shots for s in stats]
h = [s.json_metadata['d'] for s in stats]
```

```python
import seaborn as sns
```

```python
sns.set_theme(style='darkgrid')
ax = sns.lineplot(x = x, y = y, hue = h, style = h, palette=sns.color_palette('tab10')[:4], markers=['o','d','X','s'], dashes=False);
ax.set(xlabel='Rondas', ylabel='Fidelidad');
#ax.set(yscale='log');
```

```python
X = [s.json_metadata['d'] for s in stats if s.json_metadata['r'] == 1]
Y = [s.errors / s.shots for s in stats if s.json_metadata['r'] == 1]
```

```python
ax = sns.lineplot(x = X, y = Y, style=['1 ronda']*4, markers=['D']);
ax.set(xlabel='Distancia', ylabel='Tasa de fallo');
#ax.set(yscale='log')
```

```python
sns.reset_orig()
# matplotlib.rc_file_defaults()
fig, ax = plt.subplots(1, 1)
sinter.plot_error_rate(
    ax=ax,
    stats=stats,
    x_func=lambda stats: stats.json_metadata['r'],
    #failure_units_per_shot_func=lambda stats: 1 - stats.errors / stats.shots,
    group_func=lambda stats: stats.json_metadata['d']
)
ax.set_yscale('log')
ax.set_title('Surface Code Logical Error Rate')
ax.set_xlabel('Rounds')
ax.set_ylabel('Infidelity')
ax.grid(which='major')
ax.grid(which='minor')
ax.legend();
#fig.set_dpi(120)
```
